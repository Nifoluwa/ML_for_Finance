```python
# Import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import warnings
warnings.filterwarnings(action="ignore")
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.arima.model import ARIMAResultsWrapper
import statsmodels.graphics.tsaplots as sgt
from scipy.stats.distributions import chi2
import seaborn as sns
sns.set()

# Read data into memory
raw_csv_data = pd.read_csv("Index2018.csv")
df_comp = raw_csv_data.copy()

# Data cleaning
df_comp.date = pd.to_datetime(df_comp.date, dayfirst=True)
df_comp.set_index("date", inplace=True)
df_comp = df_comp.asfreq('b')
df_comp.spx.fillna(method="ffill", inplace=True)
df_comp.ftse.fillna(method="bfill", inplace=True)
df_comp.dax.fillna(value=df_comp.dax.mean(), inplace=True)
df_comp.nikkei.fillna(method="ffill", inplace=True)
df_comp["market_value"] = df_comp.ftse
del df_comp["dax"], df_comp["ftse"], df_comp["nikkei"], df_comp["spx"]

# Split data into training and test sets
size = int (len(df_comp) * 0.8)
df = df_comp.iloc[:size, :]
df_test = df_comp.iloc[size:, :]

# Create returns column for computation purposes
df["returns"] = df.market_value.pct_change().mul(100)
#df = df.iloc[1:]
```

### The Log-likelihood ratio test.
#### The likelihood ratio test is a test of the goodness of fit of models and comparing them. It checks if adding parameters to a model significantly improves it's performance and is used to evaluate model selections in time series analysis

```python
# LLR _ Test
def LLR_test(mod_1, mod_2, df=1):
    L1 = mod_1.fit().llf
    L2= mod_2.fit().llf
    LR = (2*(L2-L1))
    p = chi2.sf(LR, df).round(3)
    return p
```

#### Notes on Model Selection

##### In general, When selecting time series models, the nature of the series determines the model to be chosen. In general, AR(Auto-regressive models), MA(moving average) and ARMA(Auto-regressive Moving average) models are best for stationary series while ARIMA(Auto-Regressive Integrated Moving Average) models are best for non-stationary series. Python provides inbuilt summary and graphical tools to perform these checks from it's statsmodels class

##### Notes on ARIMA

- **ARIMA stands for Auto Regessive Integrated Moving Average. The model's goal is to predict future values of a selected series by examining the differences between values in the series instead of through the actual values. An ARIMA model can be understood by outlining each of its components as follows:**

- **Auto Regression (AR):** Refers to a model that shows a changing variable that regresses on its own lagged or prior values.

- **Integrated (I):** Represents the differencing of raw observations to allow for the time series to become stationary (i.e., data values are replaced by the difference between the data values and the previous values).

- **Moving Average (MA):** Incorporates the dependency between an observation and a residual error from a moving average model applied to lagged observations.

#### AD-Fuller Test for stationarity

###### The adfuller test provides information on the nature of the series:
###### In general, all null (H0) hypotheses are the 'no' answers to hypotheses questions while H1 is the alternative.
###### The pertinent questions in modelling are:
###### 1) Is the series non-stationary or not?
###### 2) Are the co-efficients of model parameters not significantly different from 0?

###### For chosen levels of significance (primarily at 0.05 level), based on computational outputs, decisions on model feasibility are made.


##### Techniques to convert non-stationary series to stationary series however, do exist. The differences in periods, instead of the actual period values can be computed in what can be best desribed as 'brute force' stationarity, these allow for the application of AR, MA and ARMA models before the application of more complex ARIMA models. The reasoning behind this approach is thus:
###### 1) In general, Simpler models are preferred to and should always be chosen over more complex ones.
###### 2) They offer insights into expected performances from the more complex ARIMA models

```python
sts.adfuller(df.market_value)
```
